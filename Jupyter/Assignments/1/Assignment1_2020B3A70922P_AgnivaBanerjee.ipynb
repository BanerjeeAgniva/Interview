```python
import pandas as pd
df=pd.read_csv("pincode.csv")
```

```python
rows=len(df.axes[0])
columns=len(df.axes[1])
print('Number of rows' , rows)
print('Number of columns' , columns)
df.shape
```

```python
df1=df.isna().any(axis=0)
# df1
print('Number of columns with missing values')
print(df1.sum())
df1=df.isna().any(axis=1)
#df1
print('Number of rows with missing values')
print(df1.sum())
```

```python
df.dtypes
```

```python
convert={'Latitude' : str , 'Longitude' : str}
df=df.astype(convert)
df.dtypes
```

```python
df=df.dropna(axis='rows')
# df
# df=df.reset_index(drop=True)
print('Number of rows and columns after dropping rows with NA')
print(len(df.axes[0]),'    ',len(df.axes[1]))
df.size
```

```python
convert={'Latitude' : float , 'Longitude' : float}
df=df.astype(convert)
df.dtypes
```

```python
df.astype(convert,errors="ignore")
df.dtypes
```

```python
def extractFloat(s):
    def f(c):
      return c.isdigit() or c=="."
    s1=filter(f,s)
    return "".join(s1)
df["Latitude"]=df["Latitude"].apply(extractFloat)
df["Longitude"]=df["Longitude"].apply(extractFloat)
df
```
