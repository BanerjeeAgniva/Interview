```python
import nsepython as nse
```

```python
symbol = "NIFTY BANK"
start_date = "01-May-2014"
end_date = "10-Sep-2023"
nse.index_history(symbol,start_date,end_date)
```

```python
dfbank=nse.index_total_returns(symbol,start_date,end_date)
symbol="NIFTY 50"
df50=nse.index_total_returns(symbol,start_date,end_date)
symbol="NIFTY IT"
dfit=nse.index_total_returns(symbol,start_date,end_date)
```

```python
dfit['TotalReturnsIndex']=dfit['TotalReturnsIndex'].astype(float)
df50['TotalReturnsIndex']=df50['TotalReturnsIndex'].astype(float)
dfbank['TotalReturnsIndex']=dfbank['TotalReturnsIndex'].astype(float)
```

```python
dfit
```

```python
dfit['Returns']=100*(dfit['TotalReturnsIndex']-dfit['TotalReturnsIndex'].iloc[-1])/dfit['TotalReturnsIndex'].iloc[-1]
```

```python
dfit
```

```python
df50['Returns']=100*(df50['TotalReturnsIndex']-df50['TotalReturnsIndex'].iloc[-1])/df50['TotalReturnsIndex'].iloc[-1]
```

```python
dfbank['Returns']=100*(dfbank['TotalReturnsIndex']-dfbank['TotalReturnsIndex'].iloc[-1])/dfbank['TotalReturnsIndex'].iloc[-1]
```

```python
df50
```

```python
dfbank
```

```python
frames=[dfbank,df50,dfit]
dffinal=pd.concat(frames)
dffinal
```

```python
dffinal['Index Name']=dffinal['Index Name'].replace({'NIFTY BANK' : 'Nifty Bank'})
dffinal['Index Name']=dffinal['Index Name'].replace({'NIFTY 50' : 'Nifty 50'})
dffinal['Index Name']=dffinal['Index Name'].replace({'NIFTY IT' : 'Nifty IT'})
dffinal['Date']=pd.to_datetime(dffinal['Date'],format="%d %b %Y")
dffinal.sort_values(by=['Date'],inplace=True)
dffinal=dffinal.set_index('Index Name')
dffinal
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(20,8)})
plot=sns.lineplot(x="Date",y="Returns",hue="Index Name",data=dffinal).set(title="Percentage Change In Index Funds Since MAY 2014")
sns.set(style='dark')
plt.title("Percentage Change In Index Funds Since MAY 2014",fontsize=20)
plt.xlabel("Date",fontsize=10)
plt.ylabel("Returns",fontsize=10)
plot
```
