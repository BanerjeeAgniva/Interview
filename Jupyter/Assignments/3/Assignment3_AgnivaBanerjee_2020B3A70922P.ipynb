```python
import requests
r = requests.get("https://data.telangana.gov.in/api/1/metastore/schemas/dataset/items/62a1cc18-7613-460b-a0b1-4b71c78fa1ce")
dataInfo = r.json()
dataInfo
```

```python
import pandas as pd
df_max = pd.read_csv(dataInfo['distribution'][0]["downloadURL"])
df_min = pd.read_csv(dataInfo['distribution'][1]["downloadURL"])
```

```python
df_min
```

```python
df_max
```

```python
df_max.rename(columns={'Mandal':'Dmcode'} , inplace=True)
```

```python
# mask = df_max['District'].values=='Medchal-Malkajgiri' & df_max['Mandal.1'].values=='Kapra'
df2=df_max.query(' District=="Medchal-Malkajgiri" ')
df2=df2.rename(columns = {'Mandal.1':'Mandal'})
df2=df2.query('Mandal =="Kapra" ')
df2
```

```python
df3=df_min.query(' District=="Medchal-Malkajgiri" ')
df3=df3.rename(columns = {'Mandal.1':'Mandal'})
df3=df3.query('Mandal =="Kapra" ')
df3
```

```python
frames=[df2,df3]
df_final=pd.concat(frames)
df_final
```

```python
df_final=df_final.transpose()
df_final
```

```python
df_final=df_final.drop('District')
df_final=df_final.drop('Mandal')
df_final
```

```python
cols=list(df_final.columns)
cols
cols[0]='Tempmax'
cols[1]='Tempmin'
df_final.columns=cols
```

```python
df_final
```

```python
df_final=df_final.drop('Dmcode')
```

```python
df_final
```

```python
df_final=df_final.reset_index()
```

```python
df_final
```

```python
df_final=df_final.rename(columns={'index' : 'Date'})
```

```python
df_final
```

```python
df_final["Date"]=pd.to_datetime(df_final["Date"],format="%b-%y")
```

```python
df_final
```

```python
df_final=df_final.set_index("Date")
```
